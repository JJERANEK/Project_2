```python
import pandas as pd
import json
```

```python
with open('../DataScraping/ticker_categories.json', 'r') as json_file:
    companies = json.load(json_file)
```

```python
companies
```

```python
category_list = []
category_dict = {}
for company in companies:
    category_dict['Ticker'] = company['ticker']
    category_dict['Company Name'] = company['name']
    del company['ticker']
    del company['name']
    category_dict['Categorical Data'] = company
    category_list.append(category_dict)
    category_dict = {}
```

```python
with open('../CleanedData/ticker_categories_cleaned.json', 'w') as new_file:
    json.dump(category_list, new_file)
```
